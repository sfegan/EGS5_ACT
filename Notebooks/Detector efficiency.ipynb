### Mirror reflectivity

```python
meff = EGS5Simulations.TelescopeEfficiency()
meff.scaleEffFromFile('../Parameters/corsika_mirreff.dat')
xmin = meff.xmin()
xmax = meff.xmax()
x = arange(xmin,xmax,(xmax-xmin)*0.001)
y = array(list(map(meff.y, x)))
plot(x,y)
a=list(axis())
a[2]=0
a[3]=1
axis(a)
grid()
xlabel('Photon energy [eV]')
ylabel('Reflectivity');
savefig('eff_mirror.pdf')
```

### Quantum efficiency

```python
qeff = EGS5Simulations.TelescopeEfficiency()
qeff.scaleEffFromFile('../Parameters/corsika_quanteff.dat')
xmin = qeff.xmin()
xmax = qeff.xmax()
x = arange(xmin,xmax,(xmax-xmin)*0.001)
y = array(list(map(qeff.y, x)))
plot(x,y)
a=list(axis())
a[2]=0
a[3]=1
axis(a)
grid()
xlabel('Photon energy [eV]')
ylabel('Quantum efficiency');
savefig('eff_qe.pdf')
```

### Atmospheric transmission

```python
atmabs = EGS5Simulations.AtmosphericAbsorption('../Parameters/corsika_atmabs.dat')
for z in arange(1.25,20.0,2.5):
    atmeff = atmabs.absorptionForAltitude(z*100000)
    xmin = atmeff.xmin()
    xmax = atmeff.xmax()
    x = arange(xmin,xmax,(xmax-xmin)*0.001)
    y = exp(-array(list(map(atmeff.y, x))))
    plot(x,y,label='z=%g km'%z)
    hold('on')
legend()
grid()
xlabel('Photon energy [eV]')
ylabel('Atmospheric transmission');
savefig('eff_atm.pdf')
```

### Integrated yield

```python
w0 = cos(30.0/180.0*pi)
w1 = w0+0.05
w2 = w0-0.05
eff = EGS5Simulations.TelescopeEfficiency()
eff.scaleEffFromFile('../Parameters/corsika_mirreff.dat')
eff.scaleEffFromFile('../Parameters/corsika_quanteff.dat')
actyield = atmabs.integrateYield(0.0, w0, eff)
z = arange(0,30,0.1)
plot(z, list(map(lambda h: actyield._yield(h*100000,w0), z)),label='$zn=%.1f^\\circ$'%(arccos(w0)/pi*180))
hold('on')
plot(z, list(map(lambda h: actyield._yield(h*100000,w2), z)),label='$zn=%.1f^\\circ$'%(arccos(w2)/pi*180))
plot(z, list(map(lambda h: actyield._yield(h*100000,w1), z)),label='$zn=%.1f^\\circ$'%(arccos(w1)/pi*180))
draw()
legend()
xlabel('Altitude [km]')
ylabel('Integrated efficiency');
```

### Coefficiants of Taylor expansion

\\[ Y(h,w) \int_{\epsilon_0\}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-\left|\tau(h,\epsilon)-\tau(h_0,\epsilon)\right|/w\right) = \int_{\epsilon_0}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-X(h,\epsilon)/w\right)
\\]

\\[ Y(h,w) = \int_{\epsilon_0}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-\left|\tau(h,\epsilon)-\tau(h_0,\epsilon)\right|/w\right) 
= \int_{\epsilon_0}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-X(h,\epsilon)/w\right) \\]

\\[Y(h,w) \approx Y(h,w_0) + \left.\frac{dY}{dw}\right|_{w_0} (w-w_0) + \left.\frac{d^2Y}{dw^2}\right|_{w_0} \frac{(w-w_0)^2}{2!} \\]

<br/>

#### Linear term:

<br/>

\\[ \frac{dY}{dw} = \int_{\epsilon_0}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-X(h,\epsilon)/w\right) \frac{X(h,\epsilon)}{w^2} \\]

<br/>

#### Quadratic term:

<br/>

\\[ \frac{d^2Y}{dw^2} = \int_{\epsilon_0}^{\epsilon_1} d\epsilon R(\epsilon) \exp\left(-X(h,\epsilon)/w\right) \left[ \frac{X^2(h,\epsilon)}{w^4} - \frac{2X(h,\epsilon)}{w^3} \right]  \\]


```python
actyield = atmabs.integrateYield(0.0, w0, eff)
nxy = actyield.nXY()
xi = array(list(map(lambda i: actyield.xi(i), range(0,nxy))))
yi0 = array(list(map(lambda i: actyield.yi(i).first, range(0,nxy))))
yi1 = array(list(map(lambda i: actyield.yi(i).second, range(0,nxy))))
yi2 = array(list(map(lambda i: actyield.yi(i).third, range(0,nxy))))
figure(1, figsize=[13.5, 6])
subplot(1,3,1)
plot(xi/100000,yi0,'x-')
title('Constant term')
xlabel('Altitude [km]')
subplot(1,3,2)
plot(xi/100000,yi1,'x-')
title('Linear term')
xlabel('Altitude [km]')
subplot(1,3,3)
plot(xi/100000,2*yi2,'x-');
title('Quadratic term')
xlabel('Altitude [km]');
```
