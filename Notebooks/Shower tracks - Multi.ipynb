```python
atm = EGS5Simulations.LayeredAtmosphere('Parameters/atmprof6.dat')
ztop   = atm.topOfAtmosphere()
z0     = 0
nlayer = 100
bfield = None
nmedia = 1
emax   = 10000000
media  = EGS5Simulations.VecMedia()
layers = EGS5Simulations.VecLayer()
EGS5Simulations.EGS5AtmosphericDetector.makeMediaAndLayers(media, layers, atm, nlayer, bfield, z0, ztop, nmedia)
```

```python
det = EGS5Simulations.TrackWritingDetector(media,layers,z0,emax)
```

```python
egs5 = EGS5Simulations.EGS5System.instance()
egs5.setUI(det)
egs5.initializeEGS5()
```

```python
alle0 = [10000,100000,1000000,10000000]
f, ax = subplots(1,len(alle0),sharey=True)
for ie0 in range(0,len(alle0)):
    e0 = alle0[ie0]
    egs5.shower(e0,0,0,ztop,0,0,-1.0,layers.size()+1,0)
    det.nTracks()
    istep=10
    i=range(0,det.nTracks(),istep)
    x=list(map(lambda ii: det.track(ii).x0*1e-2, i))
    y=list(map(lambda ii: det.track(ii).y0*1e-2, i))
    z=list(map(lambda ii: det.track(ii).z0*1e-5, i))
    ax[ie0].plot(x,z,',')
    ax[ie0].axis([-500,500,0,20])
    ax[ie0].set_xlabel('Horizontal position [m]')
ax[0].set_ylabel('Altitude [km]')
f.subplots_adjust(wspace=0.1)
f.savefig('Notebooks/multi_shower.pdf')
```
