```python
%pylab inline
#matplotlib.rcParams['figure.figsize'] = 9, 6
import sys,os
sys.path.append('..')
import EGS5Simulations
from numpy import *
```

```python
C = EGS5Simulations.Cherenkov
c = C.c()*1e-9 # cm/ns
atm = EGS5Simulations.LayeredAtmosphere('../Parameters/atmprof6.dat')
ztop = 10000000 # atm.topOfAtmosphere()
z = arange(10000,ztop,0.0001*ztop)
n = array(list(map(lambda x: atm.nMinusOne(x)+1.0, z)))
```

#### Threshold

```python
threshold = array(list(map(C.threshold, n)))
loglog(z/100000,threshold)
grid()
xlabel('Altitude [km]')
ylabel('Cherenkov threshold for electrons [MeV]');
```

```python
plot(z/100000,threshold)
grid()
axis([0,30,20,160])
xlabel('Altitude [km]')
ylabel('Cherenkov threshold for electrons [MeV]');
savefig('cherenkov_threshold_30km.pdf')
```

#### Cherenkov angle and radius on ground

```python
sin2thetac = array(list(map(lambda x: C.sin2Thetac(float('inf'), x), n)))
thetac = 180.0/pi*numpy.arcsin(sqrt(sin2thetac))
semilogx(z/100000,thetac)
grid()
xlabel('Altitude [km]')
ylabel('Cherenkov angle [deg]');
savefig('cherenkov_angle.pdf')
```

```python
radius=tan(thetac/180.0*pi)*z
semilogx(z/100000,radius/100)
grid()
xlabel('Altitude [km]')
ylabel('Cherenkov radius on ground [m]');
```

#### Photon yield per unit track length (per unit detector bandwidth) and density on ground

```python
dNdL = array(list(map(lambda x: C.yieldDensity(x, 1), sin2thetac)))
semilogx(z/100000,dNdL)
grid()
xlabel('Altitude [km]')
ylabel('Photon yield [ph cm$^{-1}$ eV$^{-1}$ii]');
```

```python
ii=990
print(z[ii]/100000,dNdL[ii]*100*5)
```

```python
loglog(z/100000,dNdL/(2*pi*radius/100))
grid()
xlabel('Altitude [km]')
ylabel('Photon yield ground density [ph cm$^{-1}$ eV$^{-1}$ m$^{-1}$]');
```

#### Atmospheric time delay

```python
tc = list(map(lambda x: atm.propagationTimeCorrection(x)/c, z))
semilogx(z/100000, tc)
grid()
xlabel('Altitude [km]')
ylabel('Retardation [ns]');
```

#### Cherenkov shockwave illustration

```python
f = figure(figsize=[6.2,4.2],frameon=False)
ax = f.add_axes([0., 0., 1., 1.])
n1 = 1.33
st = sin(frange(0,360.1,1)/180*pi)
ct = cos(frange(0,360.1,1)/180*pi)
xmax = 3
ymax = 2.1
xall = frange(0,100*xmax,0.4)
for x in xall:
    plot((xmax-x)+ct*x/n1,st*x/n1,'k',alpha=xmax/max(xmax,x))
arrow(-10,0,13.5,0,color='b',width=0.01)
plot(xmax-xall,xall*0,'.b')
plot([xmax-ymax*n1*sqrt(1-1/n1**2),xmax,xmax-ymax*n1*sqrt(1-1/n1**2)],[ymax,0,-ymax],'r',lw=2)
axis('equal')
axis('off')
axis([-2.1,4.1,-2.1,2.1])
f.savefig('cherenkov_illustration_%g.pdf'%n1);
```

#### Cherenkov cones in atmosphere

```python
figure(1, figsize=[4.5, 6])
for zi in arange(0.0,30.0,2.0):
    ni = atm.nMinusOne(zi*100000)+1.0
    s2i = C.sin2Thetac(float('inf'), ni)
    thetai = arcsin(sqrt(s2i))
    tantheta = tan(thetai)
    plot([-zi*tantheta*1000, 0, zi*tantheta*1000],[0, zi, 0],'k')
    hold('on')
a=axis()
#plot([0,0],[a[2],a[3]],'k')
xlabel('Distance on ground [m]')
ylabel('Emission altitude [km]');
savefig('cherenkov_cones_air.pdf')
```

```python
figure(1, figsize=[9, 6])
xc=130
for zi in arange(0.0,30.0,2.0):
    ni = atm.nMinusOne(zi*100000)+1.0
    s2i = C.sin2Thetac(float('inf'), ni)
    thetai = arcsin(sqrt(s2i))
    tantheta = tan(thetai)
    al = 0.3+0.7*exp(-(zi-10)**2/7**2)
    plot([xc-zi*tantheta*1000, xc, xc+zi*tantheta*1000],[0, zi, 0],'k',alpha=al,lw=3*al)
    plot([-xc-zi*tantheta*1000, -xc, -xc+zi*tantheta*1000],[0, zi, 0],'k',alpha=al,lw=3*al)
    hold('on')
a=axis()
#plot([0,0],[a[2],a[3]],'k')
xlabel('Distance on ground [m]')
ylabel('Emission altitude [km]');
savefig('cherenkov_double_cones_air.pdf')
```

```python
figure(1, figsize=[9, 6])
xc=130
zi = 0
while zi<30:
    ni = atm.nMinusOne(zi*100000)+1.0
    s2i = C.sin2Thetac(float('inf'), ni)
    thetai = arcsin(sqrt(s2i))
    tantheta = tan(thetai)
    plot([xc-zi*tantheta*1000, xc, xc+zi*tantheta*1000],[0, zi, 0],'k')
    plot([-xc-zi*tantheta*1000, -xc, -xc+zi*tantheta*1000],[0, zi, 0],'k')
    zi += 3-2.8*exp(-(zi-10)**2/10**2)
    hold('on')
a=axis()
#plot([0,0],[a[2],a[3]],'k')
xlabel('Distance on ground [m]')
ylabel('Emission altitude [km]');
savefig('cherenkov_double_cones_air_2.pdf')
```
