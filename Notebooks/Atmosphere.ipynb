```python
%pylab inline
matplotlib.rcParams['figure.figsize'] = 9, 6
import sys,os
sys.path.append('..')
import EGS5Simulations
```

```python
c = EGS5Simulations.Cherenkov.c()*1e-9 # cm/ns
atm = EGS5Simulations.LayeredAtmosphere('../Parameters/atmprof6.dat')
```

```python
ztop = 10000000 # atm.topOfAtmosphere()
z = numpy.arange(0,ztop,0.001*ztop)
```

#### Density

```python
rho = list(map(lambda x: atm.rho(x), z))
semilogy(z/100000, rho)
grid()
xlabel('Altitude [km]')
ylabel('Density [g cm$^{-3}$]');
```

#### Scale height

```python
H = -1/(diff(log(rho))/diff(z))/100000
plot(z[0:-1]/100000,H)
grid()
xlabel('Altitude [km]')
ylabel('Scale height [km]');
```

#### Thickness

```python
th = list(map(lambda x: atm.thickness(x), z))
semilogy(z/100000, th)
grid()
xlabel('Altitude [km]')
ylabel('Thickness [g cm$^{-2}$]');
```

```python
plot(z/100000, th)
axis([0,20,0,1100])
xticks(frange(0,20,1))
yticks(frange(0,1100,100))
grid()
xlabel('Altitude [km]')
ylabel('Thickness [g cm$^{-2}$]');
```

#### Thickness in radiation lengths

```python
thrl = list(map(lambda x: atm.thickness(x)/36.6, z))
plot(z/100000, thrl)
grid()
xlabel('Altitude [km]')
ylabel('Radiation lengths [1]');
```

```python
ii=78; print(z[ii]/100000,thrl[ii])
ii=85; print(z[ii]/100000,thrl[ii])
ii=72; print(z[ii]/100000,thrl[ii])
log(1e12/36.7e6)
```

#### Primary energy vs shower max altitude

```python
e0 = list(map(lambda x: exp(atm.thickness(x)/36.6)*84.2E6, z))
semilogy(z/100000, e0)
grid()
xlabel('Altitude [km]')
ylabel('Primary energy [eV]');
```

#### Combo plot

```python
f = figure()
ax1 = subplot2grid([1,4],[0,0],1,1)
ax1.plot(asarray(rho)/max(rho),z/100000)
ax1.axis([0,1.25e-3,0,20])
ax1.set_xticks(frange(0,1,0.2))
ax1.set_yticks(frange(0,20,2))
ax1.grid()
ax1.set_ylabel('Altitude [km]')
ax1.set_xlabel('Relative density');
ax2 = subplot2grid([1,4],[0,1],1,1,sharey=ax1)
ax2.plot(thrl,z/100000)
ax2.axis([0,30,0,20])
ax2.grid()
ax2.set_xlabel('Thickness [$X_0$]');
ax3 = subplot2grid([1,4],[0,2],1,2,sharey=ax1)
ax3.plot(log10(asarray(e0)),z/100000)
ax3.grid()
ax3.axis([8,21,0,20])
ax3.set_xticks(frange(8,21,1))
ax3.set_xlabel('Primary energy  [$\log_{10}(E_0/eV)$]');
ax3.text(9,2.5,'1 GeV',horizontalalignment='center',verticalalignment='top',rotation=270)
ax3.text(12,2.5,'1 TeV',horizontalalignment='center',verticalalignment='top',rotation=270)
ax3.text(15,2.5,'1 PeV',horizontalalignment='center',verticalalignment='top',rotation=270)
ax3.text(18,19.5,'1 EeV',horizontalalignment='center',verticalalignment='top',rotation=270)
ax3.text(20.8,19.5,'1 ZeV',horizontalalignment='right',verticalalignment='top',rotation=270)
plt.setp([a.get_yticklabels() for a in f.axes[1:]], visible=False);
f.subplots_adjust(wspace=0.12)
f.savefig('atm_multi.pdf')
```

#### Refractive index (n-1)

```python
nmo = list(map(lambda x: atm.nMinusOne(x), z))
semilogy(z/100000, nmo)
grid()
xlabel('Altitude [km]')
ylabel('n-1');
```

#### Retardation of light

```python
tc = list(map(lambda x: atm.propagationTimeCorrection(x)/c, z))
semilogx(z/100000, tc)
grid()
xlabel('Altitude [km]')
ylabel('Retardation [ns]');
```
